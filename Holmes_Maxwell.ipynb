```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create a simple data frame

data = {
    "Student": ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [88, 72, 95, 65, 78],
    'Science': [90, 85, 92, 70, 80],
    'English': [85, 78, 88, 60, 75]
}

df = pd.DataFrame(data)
df.head()
```

```python
#calculate average score
df['Average'] = np.mean(df[['Math', 'Science', 'English']].values, axis=1)
df.head()
```

```python
#plotting bar graph of average scores
plt.figure(figsize=(8, 5))
plt.bar(df['Student'], df['Average'], color='skyblue')
plt.title("Average score of Students")
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.ylim(0, 100) #range between
plt.grid(axis='y', linestyle='--', alpha=0.9)

#show exact values above bars
for index, value in enumerate(df['Average']):
  plt.text(index, value + 1, f'{value:.1f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
#plotting multiple subjects line graph
plt.figure(figsize=(10, 6))
plt.plot(df['Student'], df['Math'], label='Math', marker='o')
plt.plot(df['Student'], df['Science'], label='Science', marker='s')
plt.plot(df['Student'], df['English'], label='English', marker='^')

plt.title("Subject wise scores of Students")
plt.xlabel("Student")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
