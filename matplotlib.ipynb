```python
#Matplotlib is a popular Python plotting library used for data visualization. It allows you to create static, animated, and interactive plots.

import matplotlib.pyplot as plt
import numpy as np
```

```python
#example 1: Simple line plot

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
#Scatter plot
x = [5, 7, 8, 7, 2, 17]
y = [99, 86, 87, 88, 100, 86]

plt.scatter(x, y, color="red")
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#bar chart

x=["A", "B", "C", "D"]
y=[10, 20, 15, 30]

plt.bar(x, y, color="green")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#pie chart

labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.axis('equal') #equal aspect ratio
plt.show()
```

```python
#histogram
import numpy as np
data = np.random.randn(100)

plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#double line plot

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
#customing plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y, color='red', linestyle="--", marker="o", linewidth=2)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid(True)
plt.show()
```

```python
#multiple lines
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.plot(x, y1, label="First Line")
plt.plot(x, y2, label="Second Line")
plt.legend()
plt.show()
```

```python
#subplots

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Subplot 1")

plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.title("Subplot 2")

plt.tight_layout()
plt.show()
```
