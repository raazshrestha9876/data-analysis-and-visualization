```python
df = pd.read_csv('market-sales-data.csv')
df.head(20)
```

```python
#question 1: sales per city

city_sales = df.groupby('City')['Total'].sum()
city_sales.plot(kind='bar', color='skyblue', title="Total Sales per City")
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.ylim(0, 2000)
plt.show()
```

```python
#question 2: Product Distribution

product_counts = df['Product'].value_counts()
product_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, title="Product Distribution")
plt.ylabel('')
plt.show()
```

```python
#question 3: Gender vs sales
gender_sales = df.groupby('Gender')['Total'].sum()
gender_sales.plot(kind='bar', color=['skyblue', 'lightcoral'], title="Total Sales by Gender")
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
#question 4: Branch wise average sales
branch_avg_sales = df.groupby('Branch')['Total'].mean()
branch_avg_sales.plot(kind='bar', color='green', title="Average Sales per Branch")
plt.xlabel('Branch')
plt.ylabel('Average Sales')
plt.ylim(0, 250)
plt.show()
```

```python
#question 5: best-selling product categories
product_sales = df.groupby('Product')['Total'].sum()
product_sales.plot(kind='bar', color='purple', title="Total Sales per Product Category")
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Question 6: Average spending by customer type (Member vs Normal)

avg_spending = df.groupby('Customer_Type')['Total'].mean()

avg_spending.plot(kind='bar', color=['orange', 'purple'], title='average spending by customer type')
plt.ylabel('Average spending')
plt.ylim(0, 250)
plt.grid(axis='y', linestyle='dotted', alpha=0.9)


for index, value in enumerate(avg_spending):
  plt.text(index, value + 1, f'{value:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
#question 7: Sales Trend over time(using Date)

daily_sales = df.groupby('Date')['Total'].sum()
daily_sales.plot(kind='line', marker='o', color='blue', title='Sales Trend Over time')
plt.ylabel("Total Sales")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

```python
#question 8: city wish branch performance

city_branch_sales = df.groupby(['City', 'Branch'])['Total'].sum()

city_branch_sales.plot(kind='bar', figsize=(8, 5), title='City-Wise Branch Performance')
plt.ylabel('Total sales')
plt.grid(axis='y', linestyle='dotted', alpha=0.9)
plt.xticks(rotation=0)
for index, value in enumerate(city_branch_sales):
  plt.text(index, value, f'{value:.2f}', ha='center')
plt.show()
```

```python
#question 9: Most profitable product category by 20 %

profit_margin = 0.20
df['Profit'] = df['Total'] * profit_margin

df.head()
```

```python
profit_by_product = df.groupby('Product')['Profit'].sum().sort_values(ascending=False)

profit_by_product.plot(kind='bar', color='darkgreen', title="Most profitable product categories")
plt.ylabel('Total Profit')
plt.grid(axis='y', linestyle='dotted', alpha=0.9)
for index, value in enumerate(profit_by_product):
  plt.text(index, value + 1, f'{value:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
